```python
# importing required packages
import seaborn as sns
import matplotlib.pyplot as plt
# loading dataset
data = sns.load_dataset("titanic") 
data.head(15)
```

```python
# draw jointplot with
# hex kind
sns.jointplot(x = "age", y = "survived",
              kind = "hex", data = data, dropna = True, palette = "autumn")
sns.set_style('darkgrid')
# show the plot
plt.show()
```

```python
#creating a Dataset
feline_family =["Cat","Tiger","Leopard","Cheetah","Lion"]
Data = [41,34,27,18,20]
explode = (0.2,0,0,0,0) # only explode the first slice ie "Cat"
mycolors = ["black", "hotpink", "b","#17becf","#4CAF50"]
fig1,ax1 = plt.subplots()
ax1.pie(Data, explode=explode,labels=feline_family, colors = mycolors, autopct ='%1.0f%%',shadow =True,startangle = 90,wedgeprops ={'edgecolor':'black'}) 
# create a pie chart
fig = plt.figure(figsize = [10,7])
ax1.axis("equal")# equal aspect ensures that pie is drawn as a circle
#show plot

ax1.legend(title = "feline_family:")
plt.show()
```
